```python
import pandas as pd
import numpy as np
import scipy
```

```python
from scipy.stats import ttest_rel
from scipy.stats import wilcoxon
```

```python
import statsmodels
from statsmodels.stats.contingency_tables import mcnemar
```

```python
import matplotlib.pyplot as plt
```

```python
df_unclean = pd.read_csv("../commed/interfinal2.csv")
df_unclean.drop(columns='Timestamp', inplace=True)
df_unclean
```

```python
df_unclean.replace('ทำ',1,inplace=True)
df_unclean.replace('ไม่ทำ',0,inplace=True)
df_unclean
```

# Poster ล้างมือ

```python
df_wash = df_unclean.iloc[:,2:6].copy()

df_wash.replace('ทำเป็นประจำ',2,inplace=True)
df_wash.replace('ทำบางครั้ง',1,inplace=True)
df_wash.replace('ไม่เคยทำ',0,inplace=True)
df_wash
```

***
# Slide**
# เทียบการล้างมือ ก่อน vs หลังเห็นโปสเตอร์
# ใช้ McNemar
# p-value = 0.125

```python
ppwash = pd.crosstab(df_wash.iloc[:,2],df_wash.iloc[:,3])
ppwash
```

```python
print (mcnemar(ppwash))
```

### ทำ Wilcoxon เผื่อไว้ p-value = 0.003 ไม่ต้องใส่ในสไลด์

```python
pre_wash = df_wash.iloc[:,2]
post_wash = df_wash.iloc[:,3]
wilcoxon(pre_wash,post_wash)
```

# Pre Post test

```python
df_pretest_wear = df_unclean.iloc[:,16:23]
df_pretest_wear
```

```python
df_pretest_discard = df_unclean.iloc[:,32:38]
df_pretest_discard
```

```python
df_pretest_weardiscard = df_pretest_wear.join(df_pretest_discard)
df_pretest_weardiscard
```

```python
pre_weardiscard = df_pretest_weardiscard.sum(axis=1)
pre_weardiscard
```

```python
df_pretest_keep = df_unclean.iloc[:,23:32]
df_pretest_keep
```

```python
pre_keep = df_pretest_keep.sum(axis=1)
pre_keep
```

```python
df_posttest_wear = df_unclean.iloc[:,38:45]
df_posttest_wear
```

```python
df_posttest_discard = df_unclean.iloc[:,54:60]
df_posttest_discard
```

```python
df_posttest_weardiscard = df_posttest_wear.join(df_pretest_discard)
df_posttest_weardiscard
```

```python
post_weardiscard = df_posttest_weardiscard.sum(axis=1)
post_weardiscard
```

```python
df_posttest_keep = df_unclean.iloc[:,45:54]
df_posttest_keep
```

```python
post_keep = df_posttest_keep.sum(axis=1)
post_keep
```
